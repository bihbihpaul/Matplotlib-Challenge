```python
# Dependencies and Setup
import os
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import random
import numpy as np
from scipy.stats import linregress
```

```python
# Reading mouse_metadata file

csv_file = os.path.join("resources","Mouse_metadata.csv")

Raw_mouse_metadata_df = pd.read_csv(csv_file)

Raw_mouse_metadata_df
```

```python
# Reading study_results file

csvfile = os.path.join("resources","Study_results.csv")

Raw_study_results_df = pd.read_csv(csvfile)

Raw_study_results_df
```

```python
# Combining the data into a single dataset

merged_data_df = pd.merge(Raw_mouse_metadata_df, Raw_study_results_df,how='outer',on='Mouse ID')

merged_data_df
```

```python
# Mouse ID with unique value
uniq_mouse_id_df = len(merged_data_df["Mouse ID"].unique())
print(f"There are:{uniq_mouse_id_df} unique Mouse IDs")
```

```python
# Checking Mouse ID duplicate with  Timepoint

dupli_mice_ID_df = merged_data_df[merged_data_df.duplicated(['Mouse ID', 'Timepoint'])]

dupli_mice_ID_df
```

```python
# Selecting all the duplicates to be removed
dupli_mouse_ID = merged_data_df.loc[merged_data_df["Mouse ID"]== "g989"]

dupli_mouse_ID
```

```python
# counting the number of duplicates
dupli_count = len(dupli_mouse_ID)
print(f"There are {dupli_count} duplicates to remove")
```

```python
# Removing the duplicate mouse by ID.
Dupli_Free_merged_data_df = merged_data_df.loc[merged_data_df['Mouse ID'] != 'g989']

Dupli_Free_merged_data_df
```

```python
# Checking the number of mice in the clean DataFrame.

number_of_mice = Dupli_Free_merged_data_df["Mouse ID"].nunique()

print(f"There are {number_of_mice} mice in the cleaned dataframe ")
```

```python
# creating a clean DataFrame
clean_data = Dupli_Free_merged_data_df
```

```python
# Grouping by Drug Regimen

grouped_drug = clean_data.groupby("Drug Regimen")

# Using groupby Drug Regimen and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

summary_stats = pd.DataFrame()
summary_stats["Mean Tumor Volume"] = grouped_drug["Tumor Volume (mm3)"].mean().round(3)
summary_stats["Median"] = grouped_drug["Tumor Volume (mm3)"].median().round(3)
summary_stats["Variance"] = grouped_drug["Tumor Volume (mm3)"].var().round(3)
summary_stats["STD Deviation"] = grouped_drug["Tumor Volume (mm3)"].std().round(3)
summary_stats["STD Error of Mean"] = grouped_drug["Tumor Volume (mm3)"].sem().round(3)
summary_stats
```

```python
# Summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

summary_table_aggr = grouped_drug.agg({"Tumor Volume (mm3)" : ["mean","median","var","std","sem"]})

summary_table_aggr
```

```python

# Calculating the final tumor volume of each mouse across four of the treatment regimens:Capomulin, Ramicane, Infubinol,Ceftamin
potential_treatment_df = clean_data.set_index("Drug Regimen").loc[["Capomulin","Ramicane", "Infubinol","Ceftamin"]].reset_index()

# Getting the last (greatest) timepoint for each mouse
promising_group = potential_treatment_df.groupby('Mouse ID')['Timepoint'].max()
promising_group_df = pd.DataFrame(promising_group)
promising_group_df

# Merging promising_group with potential_treatment dataframe to get the tumor volume at the last timepoint
regimen_df = pd.merge(promising_group_df, potential_treatment_df, on=['Timepoint', 'Mouse ID'])
regimen_df.head(100) 
```

```python
# Putting treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin","Ramicane", "Infubinol","Ceftamin"]

# Creating empty list to fill with tumor vol data (for plotting)
tumor_list = []

# Calculating the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments: 
    
    # Locating the rows which contain mice on each drug and get the tumor volumes
    tumor_vol = regimen_df .loc[regimen_df ["Drug Regimen"]== treatment, "Tumor Volume (mm3)"]
    tumor_list.append(tumor_vol)
    
    # adding subset 
    #computing values
    quartiles = tumor_vol.quantile([.25,.5,.75])
    lower = round(quartiles[.25], 4)
    upper = round(quartiles[.75], 4)
    iqr = round(upper - lower, 4)
    
    #printing values
    print(f"Drug Name: {treatment}")
    print(f"Lower Quartile: {lower}")
    print(f"Lower Quartile: {upper}")
    print(f"Interquartile range: {iqr}")
        
    # Determining outliers using upper and lower bounds
    #calculating values
    lower_b = round(lower - (iqr*1.5), 4)
    upper_b = round(upper + (iqr*1.5), 4)
    
    #printing values
    print(f"Values below may be outliers: {lower_b}")
    print(f"Values below may be outliers: {upper_b}")
    print("\n")
```

```python
# Generating a box plot of the final tumor volume of each mouse across promising regimens 
# setting y axis labels
Potential_Regimen = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# formatting box plot
fig, ax1 = plt.subplots(figsize=(15,10))
ax1.set_title("Final Tumor Volume per Regimen")
ax1.set_ylabel("Tumor Volume (mm3)")
ax1.set_xlabel("Potential Regimen")

# marking outliers
flierprops = dict(marker="o", markerfacecolor="r", markersize=10)

ax1.boxplot(tumor_list, labels=Potential_Regimen, flierprops=flierprops,patch_artist=True, widths=0.4)
plt.savefig("Images/Tumor Size by Drug Regimen.png", bbox_inches="tight")
plt.show()
```

```python
# Generating a scatter plot of mouse weight versus average tumor volume for the Capomulin treatment regimen.

fig1, ax1 = plt.subplots(figsize=(15, 10))
avg_capomu_vol =capomu_mice_df.groupby(["Mouse ID"]).mean()

marker_size=15
plt.scatter(avg_capomu_vol["Weight (g)"],avg_capomu_vol["Tumor Volume (mm3)"],s=175, color="blue")
plt.title("Mouse Weight Versus Average Tumor Volume",fontsize =25)
plt.xlabel("Weight (g)",fontsize =14)
plt.ylabel("Averag Tumor Volume (mm3)",fontsize =14)


plt.savefig("Images/scatterplot.png", bbox_inches = "tight")
plt.show()
```

```python
# Computing the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr=round(st.pearsonr(avg_capomu_vol["Weight (g)"],avg_capomu_vol["Tumor Volume (mm3)"])[0],2)
print(f"The correlation between mouse weight and average tumor volume is {corr}")
```

```python
x_values = avg_capomu_vol["Weight (g)"]
y_values = avg_capomu_vol["Tumor Volume (mm3)"]

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

print(f"slope:{slope}")
print(f"intercept:{intercept}")
print(f"rvalue (Correlation coefficient):{rvalue}")
print(f"pandas (Correlation coefficient):{corr}")
print(f"stderr:{stderr}")

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(line_eq)
```

```python
# Adding the linear regression equation and line to plot
fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.scatter(x_values,y_values,s=175, color="blue")
plt.plot(x_values,regress_values,"r-")
plt.title("Regression Plot of Mouse Weight Versus Average Tumor Volume",fontsize =20)
plt.xlabel("Weight(g)",fontsize =14)
plt.ylabel("Average Tumor Volume (mm3)",fontsize =14)
ax1.annotate(line_eq, xy=(20, 40), xycoords="data",xytext=(0.8, 0.95), textcoords="axes fraction",horizontalalignment="right", verticalalignment="top",fontsize=30,color="g")

print(f"The r-squared is: {rvalue**2}")

plt.savefig("Images/linear_regression.png", bbox_inches = "tight")
plt.show()
```
